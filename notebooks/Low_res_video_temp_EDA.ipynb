```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
import os

arr = os.listdir("../csv_data")
```

```python
data = list()
for i in arr:
    data.append(pd.read_csv("../csv_data/"+i))
```

```python
data[2]
```

```python
stages = ["stage1","stage2","stage3","stage4","stage5","stage6"]
df = pd.concat([x["temp"] for x in data], axis=1)
```

```python
df.columns = stages
```

```python
stage = 0
for i in data:
    plt.plot(i.index,i['temp'], label= stage +1)
    stage = stage +1

plt.legend()
plt.show()
```

Handling missing values

```python
df.fillna(method = "ffill").plot()
```

```python
df.fillna(method = "bfill").plot()
```

```python
df.fillna(df.mean()).plot()
```

```python
dataframe = df.fillna(method = "bfill")
df_melt = dataframe.melt( value_vars=['stage1', 'stage2','stage3','stage4','stage5','stage6'])
df_melt
```
